```python
import requests

apikey='6d43785a7a64656d32355066564551'
url = f'http://openapi.seoul.go.kr:8088/{apikey}/json/energyUseDataSummaryInfo/1/5'

response = requests.get(url)
response_json = response.json()
if response.status_code == 200:
    print("api 호출 성공")
    print(response_json)
else:
    print(f"API 호출 실패: {response.status_code}")


result = []
for item in response_json["energyUseDataSummaryInfo"]["row"]:
    if item.get("MM_TYPE") == "개인":
        result.append({
            "EUS": item.get("EUS"),
            "GUS": item.get("GUS"),
            "WUS": item.get("WUS"),
            "HUS": item.get("HUS")
        })

# 결과 출력
print(result)
```

```python
import requests
import time

apikey='6d43785a7a64656d32355066564551'
url = f'http://openapi.seoul.go.kr:8088/{apikey}/json/energyUseDataSummaryInfo/1/5'

response = requests.get(url)
response_json = response.json()
if response.status_code == 200:
    print("api 호출 성공")
    print(response_json)
else:
    print(f"API 호출 실패: {response.status_code}")

# 결과 저장용 리스트
results = []

# 연도: 2015 ~ 2024
for year in range(2015, 2025):
    for month in range(1, 13):
        mm = f"{month:02d}"  # '01', '02', ..., '12'
        url = f'http://openapi.seoul.go.kr:8088/{apikey}/json/energyUseDataSummaryInfo/1/10/{year}/{mm}'

        response = requests.get(url)
        energyUseData = response.json()

        rows = energyUseData.get("energyUseDataSummaryInfo", {}).get("row", [])
        for row in rows:
            if row.get("MM_TYPE") == "개인":
                results.append({
                    "year": row.get("YEAR"),
                    "month": row.get("MON"),
                    "EUS": row.get("EUS"),
                    "GUS": row.get("GUS"),
                    "WUS": row.get("WUS"),
                    "HUS": row.get("HUS")
                })

            # 너무 빠른 호출을 피하기 위해 잠깐 대기 (서버 부하 방지)
        time.sleep(0.2)

# 결과 출력
for item in results:
    print(item)
```
